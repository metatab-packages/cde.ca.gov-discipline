```python
cm = {
 'AcademicYear': 'AcademicYear',
 'ï»¿AcademicYear': 'AcademicYear',
 'CountyCode': 'CountyCode',
 'CountyName': 'CountyName',
 'Cumulative Enrollment': 'CumulativeEnrollment',
 'CumulativeEnrollment': 'CumulativeEnrollment',
 'DistrictCode': 'DistrictCode',
 'DistrictName': 'DistrictName',
 'SchoolCode': 'SchoolCode',
 'SchoolName': 'SchoolName',
 'CharterYN': 'is_charter',
 'AggregateLevel': 'AggregateLevel',
 'ReportingCategory': 'ReportingCategory',
 'Expulsion Count Defiance-Only': 'Defiance',
 'Expulsion Count Illicit Drug-Related': 'Drugs',
 'Expulsion Count Other Reasons': 'Other',
 'Expulsion Count Violent Incident (Injury)': 'Violent Injury',
 'Expulsion Count Violent Incident (No Injury)': 'Violent No Injury',
 'Expulsion Count Weapons Possession': 'Weapons ',
 'Expulsion Rate (Total)': 'Rate',
 'Total Expulsions': 'Total Expulsions',
 'Unduplicated Count of Students Expelled (Defiance-Only)': 'Undup Count Expelled Defiance',
 'Unduplicated Count of Students Expelled (Total)': 'Undup Count Expelled'}

# Interving dict make values unique, and unline set(),  preserves order
exp_cols =  {v:k for k, v in cm.items()}.keys()

e_frames = [r.dataframe().rename(columns=cm) for r in pkg.references() if r.name.startswith('exp') ] 
expl = pd.concat(e_frames)[exp_cols]

def to_numeric(df):
    df = df.replace('*', np.nan)
    for c in df.columns:
        try:
            df[c] = pd.to_numeric(df[c])
        except ValueError:
            pass
        
    return df

expl = to_numeric(expl)
        
expl
```
